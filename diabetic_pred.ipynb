```python
import  numpy as np
import  pandas as pd
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
### import dataset
```

```python
diabetics_dataset=pd.read_csv('diabetes.csv')
```

```python
## print the 5 data rows in dataset
print(diabetics_dataset.head(5))
```

```python
##getting the statistical data
diabetics_dataset.describe()
```

```python
diabetics_dataset['Outcome'].value_counts() 
## 0-> None diabetic
##1-> diabetic
```

```python
diabetics_dataset.groupby('Outcome').mean()
```

```python
##separating the data and labels
X = diabetics_dataset.drop('Outcome', axis=1)
Y = diabetics_dataset['Outcome']
```

```python
print(X)
print(Y)
```

```python
##data standerdization
Scaler = StandardScaler().fit(X)

```

```python
standerdize_data= Scaler.transform(X)
```

```python
print(standerdize_data)
```

```python
X = standerdize_data
Y= diabetics_dataset['Outcome']
print(X)
print(Y)
```

```python
## train data split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
## training the model
classifier = svm.SVC(kernel='linear')
```

```python
##training the support vector machine clasifier
classifier.fit(X_train,Y_train)
```

```python
print('slope (m):',classifier.coef_[0][0])
print('intercept ()b:',classifier.intercept_[0])
y_pred=classifier.predict(X_test)
```

```python
plt.scatter(range(len(X_test)), Y_test, color='blue', label='Actual Result')
plt.plot(range(len(X_test)), y_pred, color='red', label='Predicted Result', linewidth=2)

plt.xlabel('Samples')
plt.ylabel('Outcome')
plt.legend()
plt.show()
```

```python
##accuracy score on the training data
X_train_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print('Accuracy score on the data')
print(training_data_accuracy)
```

```python
X_test_prediction=classifier.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print('Accuracy score on the data')
print(testing_data_accuracy)
```
